```python
median_age = clinic["age_years"].median()
print(median_age)
```

# Task 5

```python
clinic.groupby("visit_reason")["visit_cost_usd"].transform('median')
```

```python
clinic["visit_cost_usd"] = clinic["visit_cost_usd"].fillna(
    clinic.groupby("visit_reason")["visit_cost_usd"]
    .transform("median")
)
```

```python
clinic['visit_cost_usd'].isna().sum()
```

### Task 5 Question:
group-based filling is more accurate because within each category, the cost will likely be much more similar. Between categories, cost can vary more. So grouping by reason for visit allows for a more accurate estimate of the missing costs.

# Task 8 discussion questions
1. animal type is much more ambiguous than taking a median age for animals that typically don't live more than a few decades.
2. The median would have taken the median over all the different visit reasons making the data skewed towards less focused areas. Grouping took the median in each category for each service.
3. The other data is still usable and adds value. Deleting the rows would take away from our sample size in other columns.
